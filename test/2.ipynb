```python
import tkinter as tk
from tkinter import filedialog
from tkinter import ttk
import os

def select_folder():
    folder_path = filedialog.askdirectory()
    if folder_path:
        file_listbox.delete(0, tk.END)  # Clear previous file list

        # Get a list of files in the selected folder
        files = os.listdir(folder_path)
        total_files = len(files)

        for file in files:
            file_listbox.insert(tk.END, file)
            progress_bars[file] = ttk.Progressbar(root, mode='determinate')
            progress_bars[file].pack()

        # Update progress bar value and maximum
        overall_progress_bar['value'] = 0
        overall_progress_bar['maximum'] = total_files

def update_progress(file):
    # Update progress bar value
    progress_bars[file]['value'] += 1

    # Update overall progress bar value
    overall_progress_bar['value'] += 1

root = tk.Tk()

select_folder_button = tk.Button(root, text="Select Folder", command=select_folder)
select_folder_button.pack()

file_listbox = tk.Listbox(root)
file_listbox.pack(side=tk.LEFT)

overall_progress_bar = ttk.Progressbar(root, mode='determinate')
overall_progress_bar.pack(side=tk.RIGHT, fill=tk.Y)

progress_bars = {}  # Dictionary to store progress bars for each file

update_button = tk.Button(root, text="Update Progress", command=lambda: update_progress(file_listbox.get(tk.ACTIVE)))
update_button.pack()

root.mainloop()
```
