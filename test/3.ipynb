```python
import tkinter as tk
from tkinter import filedialog
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import os

root = tk.Tk()
root.title("Plagiarism Checker")

file_Names = []
file_Notes = []
vectors = []
s_vector = []
pg_Results = set()
plagiarism_results = {}

def select_folder():
    folder_path = filedialog.askdirectory()
    if folder_path:
        folder_entry.delete(0, tk.END)
        folder_entry.insert(tk.END, folder_path)
        load_files()

def load_files():
    global file_Names, file_Notes, vectors, s_vector
    folder_path = folder_entry.get()
    file_listbox.delete(0, tk.END)
    pg_Results.clear()
    plagiarism_results.clear()

    file_Names = [i for i in os.listdir(folder_path) if i.endswith('.txt')]
    file_Notes = [open(os.path.join(folder_path, data), encoding='utf-8').read() for data in file_Names]
    vectors = vectorize(file_Notes)
    s_vector = list(zip(file_Names, vectors))

    for file_name in file_Names:
        file_listbox.insert(tk.END, file_name)

def vectorize(text):
    return TfidfVectorizer().fit_transform(text).toarray()

def similarity(doc1, doc2):
    return cosine_similarity([doc1, doc2])

def check_plagiarism():
    global pg_Results, plagiarism_results
    pg_Results = set()
    for a1, a2 in s_vector:
        new = s_vector.copy()
        index = new.index((a1, a2))
        del new[index]
        for z1, z2 in new:
            s_score = similarity(a2, z2)[0][1]
            pair = sorted((a1, z1))
            score = (pair[0], pair[1], s_score)
            pg_Results.add(score)

    plagiarism_results = {}
    for i in range(len(file_Names)):
        total_score = 0
        count = 0
        for j in pg_Results:
            if file_Names[i] in j:
                for k in j:
                    if isinstance(k, float):
                        total_score += k
                        count += 1
        if count > 0:
            plagiarism_results[file_Names[i]] = int((total_score / count) * 100)

    result_text.delete(1.0, tk.END)
    for file_name, percentage in plagiarism_results.items():
        result_text.insert(tk.END, f"{file_name}: {percentage}% plagiarized\n")

root.geometry("500x400")

folder_label = tk.Label(root, text="Folder Path:")
folder_label.pack(pady=10)

folder_entry = tk.Entry(root, width=50)
folder_entry.pack()

select_folder_button = tk.Button(root, text="Select Folder", command=select_folder)
select_folder_button.pack(pady=10)

file_listbox = tk.Listbox(root, width=50, height=10)
file_listbox.pack()

check_button = tk.Button(root, text="Check Plagiarism", command=check_plagiarism)
check_button.pack(pady=10)

result_label = tk.Label(root, text="Results:")
result_label.pack()

result_text = tk.Text(root, width=50, height=10)
result_text.pack()

root.mainloop()

print(plagiarism_results)
```
